```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df.drop(['deck','embark_town','alive','class','who','adult_male'], axis= 1, inplace = True)
```

```python
df['embarked'].value_counts()
```

```python
#df['embarked'] = df['embarked'].map({'A': 1, 'B': 0, 'Q':2})
```

```python
df['sex'] = df['sex'].map({'male': 1, 'female': 0})
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df['alone'] = df['alone'].astype(int)
```

```python
df['embarked'] = df['embarked'].fillna('S')
```

```python
len(df['embarked'])
```

```python
embarked_dummies = pd.get_dummies(df['embarked'], prefix='embarked', dtype=int)
df = df.drop('embarked', axis=1)
df = pd.concat([df, embarked_dummies], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x = df.drop(['survived'], axis = 1)
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
df.head()
```

```python
mod = model.fit(x_train,y_train)
```

```python
y_prde = model.predict(x_test)
y_prde
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_prde)
```

```python
confusion_matrix(y_test, y_prde)
```

```python
print(classification_report(y_test, y_prde))
```
